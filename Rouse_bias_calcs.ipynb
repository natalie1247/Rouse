Rouse gender bias calculations

For every word in test_words, I calculate the Euclidean distance between it and each of the comparison_words. 
In my output file these are in a format so that they can be copy-pasted into excel and then further processed from there

This code needs running twice: once with the human dataset, once with the machine dataset

```python
# adjust the next 2 lines to have the desired input files

meta_file_list = ["m_meta1.tsv", "m_meta2.tsv", "m_meta3.tsv", "m_meta4.tsv", "m_meta5.tsv", "m_meta6.tsv", "m_meta7.tsv", "m_meta8.tsv", "m_meta9.tsv", "m_meta10.tsv"] 
vector_file_list = ["m_vect1.tsv", "m_vect2.tsv", "m_vect3.tsv", "m_vect4.tsv", "m_vect5.tsv", "m_vect6.tsv", "m_vect7.tsv", "m_vect8.tsv", "m_vect9.tsv", "m_vect10.tsv"]

test_words = ["llorar", "poder", "trabajar", "gana", "decidio", "matrimonio", "exito", "fuerte", "cuidar"]
comparison_words = ["mujer", "hombre", "hermana", "hermano", "hija", "hijo", "madre", "padre", "senora", "senor"]

import csv
import math

count = int(input("Which number do you want to run?")) - 1

vector_file = vector_file_list [count]
meta_file = meta_file_list[count]

biglist = []
with open(meta_file) as tsv:
  for line in csv.reader(tsv, dialect = "excel-tab"):
      biglist.append(line)

print(biglist)

new_file = open("useme.csv", "w")

comparison_list = []
      
with open(vector_file) as tsv:
  index = 0
  for line in csv.reader(tsv, dialect = "excel-tab"):
    word = biglist[index]
    if word[0] in comparison_words:
        write_line = []
        for item in line:
          write_line.append(item)
        write_line.append(word[0])
        comparison_list.append(write_line)
    index = index + 1

test_list = []

with open(vector_file) as tsv:
  index = 0
  for line in csv.reader(tsv, dialect = "excel-tab"):
    word = biglist[index]
    if word[0] in test_words:
        write_line = []
        for item in line:
          write_line.append(item)
        write_line.append(word[0])
        test_list.append(write_line)
    index = index + 1

long_list = []

for item in test_list:
  for compare in comparison_list:
    counter = 0
    process_list = []
    while counter < len(compare) - 1 :
      compare_vect = float(compare[counter])
      item_vect = float(item[counter])
      subtract_vect = compare_vect - item_vect
      squared = subtract_vect**2
      process_list.append(squared)
      counter = counter + 1
    
    sum = 0
    for number in process_list:
      sum = sum + number
    final = math.sqrt(sum)

    display = []
    display.append(compare[-1])
    display.append(item[-1])
    display.append(final)

    
    long_list.append(display)


new_file = open("am10results.tsv","w")  
for part in long_list:
  print(part)
  new_file.write('\t'.join([str(x) for x in part]) + "\n")

new_file.close()

```
